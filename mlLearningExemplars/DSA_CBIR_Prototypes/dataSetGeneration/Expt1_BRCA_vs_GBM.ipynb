```python
import girder_client
import sys
from random import shuffle
import pprint
gc= girder_client.GirderClient(apiUrl='http://candygram.neurology.emory.edu:8080/api/v1/')
gc.authenticate(interactive=True)
```

```python
for c in gc.listCollection():
    if 'TCGA' == c['name']:
        TCGA_Collection = c['name']
```

```python
### List the TCGA Cohorts

cohortInfoDict = {}
for c in  gc.get('/tcga/cohort')['data']:
    cohortInfoDict[c['name']] = c
```
